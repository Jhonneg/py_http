```python
from dotenv import load_dotenv, find_dotenv
from openai import OpenAI

load_dotenv(find_dotenv(), override=True)
client = OpenAI()
```

```python
text = "This is a test of OpenAI text to speech API, using Python notebooks"
```

```python
cost = len(text) / 1000 * 0.15  # cost in cents
cost
```

```python
response = client.audio.speech.create(model="tts-1", voice="echo", input=text)

# output

response.stream_to_file("audio/tts.mp3")
```
