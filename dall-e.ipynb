# Generating Images with DALL-E 3

```python
from dotenv import load_dotenv, find_dotenv
from openai import OpenAI

load_dotenv(find_dotenv(), override=True)
client = OpenAI()

prompt = (
    "A photorealistic image of a tuxedo cat curled up on a window"
    "gazing out a busting city street slick with rain"
)

response = client.images.generate(
    model="dall-e-3",
    prompt=prompt,
    style="vivid",
    size="1024x1024",
    quality="standard",
    n=1,
)

print(response)
```

```python
image_url = response.data[0].url
print(image_url)
```

```python
import requests
import shutil

image_resourse = requests.get(image_url, stream=True)
print(image_resourse.status_code)

if image_resourse.status_code == 200:
    image_name = "images/dalle-tuxedo-cat.png"
    with open(image_name, "wb") as f:
        shutil.copyfileobj(image_resourse.raw, f)
else:
    print("Error accessing the image!")
```
