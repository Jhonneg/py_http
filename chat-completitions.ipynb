```python
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv(), override=True)
```

```python
from openai import OpenAI

client = OpenAI()
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "developer", "content": "You are a sci-fi writer and a poet"},
        {
            "role": "user",
            "content": "Write a poem about a lonely robot exploring a futuristic city",
        },
    ],
    temperature=1,
)

print(response.choices[0].message.content)
```
